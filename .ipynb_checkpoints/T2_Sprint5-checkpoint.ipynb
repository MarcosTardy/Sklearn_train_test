Attribute Information:

    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's

El gràfic és una matriu de correlació modificada com un mapa de calor per visualitzar més fàcilment si dues variables tenen correlació o no. Quant més càlid sigui el color, més correlació existirà.

Com a exemple, si busquem la intercepció entre la variable "RAD" i la variable "TAX" es pot veure que el índex de correlació és de 0.91, per tant són proporcionals. En canvi, entre les variables "MEDV" i "LSTAT" hi ha una relació inversament proporcional.

La variable onjectiu és "MEDV" que és el preu de les vivendes ocupades en milers de dólars. Aquesta variable és la que es vol predir a partir de train-test.

Les variables que tenen correlació amb el preu de la vivenda són "RM" que és el número d'habitacions per habitatge i la variable "LSTAT" que és percentatge del menor estatus de la població. Té sentit, perquè normalment quant més alt és el preu de la vivenda, més alt és el nombre d'habitacions que té i més baix és el percentatge de la població amb menor estatus.

Com que aquestes dues variables són les úniques que ens serveixen per predir el valor de "MEDV", només assignarem a la variable "X" o "features" aquestes dues. A la variable "y" o "target" que és la variable objectiu li assignarem "MEDV". 

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
X = df.drop(["MEDV"], axis=1)
y= df[["MEDV"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape) 
```

Per saber si aplicaré estandarització, primer necessito saber si les distribucions de les variables són gaussianes. 

Per començar especificaré la hipotesis nul·la i la hipotesis alternativa.

H0: La variable segueix una distribució Normal

H1: La variable NO segueix una distribució Normal

```python
from scipy.stats import shapiro #Llibreria de python importa normaltest per fer l'arlgorisme de Shapiro.

columnes=X_train.columns.tolist() #emmagatzema una llista amb els noms de les columnes del X_train
for i in columnes: #Fa tants tests d'hip com columnes hi hagin.
    stat, p = shapiro(X_train[i]) 
    #L'algorisme normaltest assigna automàticament l'estadistica del test i el p-value.
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05: # L'alpha o nivell de significància és 0,05. Estem al 95% segurs...
     print('La variable', i,' probablement segueix una distribució Normal') #Si el p-value és major que alpha, no rebutjem la hipòtesi nula
    else:
     print('La variable ', i, ' probablement NO segueix una distribució Normal') # Si és més petit, rebutjem la hipòtesi nula.
```

```python
columnes=X_train.columns.tolist() #emmagatzema una llista amb els noms de les columnes del X_train
for i in columnes: #Fa tants tests d'hip com columnes hi hagin.
    stat, p = shapiro(X_test[i]) 
    #L'algorisme normaltest assigna automàticament l'estadistica del test i el p-value.
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05: # L'alpha o nivell de significància és 0,05. Estem al 95% segurs...
     print('La variable', i,' probablement segueix una distribució Normal') #Si el p-value és major que alpha, no rebutjem la hipòtesi nula
    else:
     print('La variable ', i, ' probablement NO segueix una distribució Normal') # Si és més petit, rebutjem la hipòtesi nula.
```

Cap de les distribucions és normal, ni del training set ni del test set. Per tant, no es pot aplicar estandarització.
